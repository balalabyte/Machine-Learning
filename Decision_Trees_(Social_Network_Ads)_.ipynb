```python
import numpy as np
import pandas as pd
```

```python
data= pd.read_csv('Social_Network_Ads.csv')
data.head()
```

```python
data.info()
```

```python
data['Gender'].replace({'Male':0, 'Female':1}, inplace=True)
```

```python
x=data.iloc[:,1:4].values
y=data.iloc[:,1:4].values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score
```

```python
#Train-test split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)
```

```python
#Fit the model on training data
#clf= DecisionTreeClassifier()
#later on correction to avoid over fitting
clf= DecisionTreeClassifier(max_depth=3)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
#plotting the tree to see how the descision is made
from sklearn.tree import plot_tree
from matplotlib.pylab import rcParams
rcParams['figure.figsize']= 80,50

plot_tree(clf)

#we can see there has been a lot of overfitting and the calculations made to arrive at the descisin tree is quite a lot
#might be a case of overfitting since the tree is quite large
#so we can change the depth of the tree  in clf object definition
```
